# Now read that data to perform the grouping

```python
import pandas as pd
df = pd.read_excel(path+"\Entries.xlsx",engine='openpyxl')
```

```python
df['Entries'] = df['Entries'].astype(int)
df = df.sort_values(by="Entries")
```

```python
df.drop(df[df['Entries'] == '\x0c'].index, inplace = True)
df.drop(df[df['Entries'] == ''].index, inplace = True) 
df.drop_duplicates(subset = ['Entries'] ,inplace = True)
data = df['Entries']
```

```python
print(df.head())
df.tail()
```

```python
df
```

```python
print("Min: "+str(df.Entries.min()))
print("Max: "+str(df.Entries.max()))
```

```python
max_length = len(str(df.Entries.max()))
max_length
```

```python
# data = [10,15,22,34,85,90,101,200,2000,2005,2010,2011,2013,40000]

# create a list of the gaps between the consecutive values
gaps = [y - x for x, y in zip(data[:-1], data[1:])]
# have python calculate the standard deviation for the gaps
# sd = stdev(gaps)
# sd = 10
sd = 20
# create a list of lists, put the first value of the source data in the first
lists = [[data[0]]]
# print(lists)
# print(lists[-1][-1])
for x in data[1:]:
    # if the gap from the current item to the previous is more than 1 SD
    # Note: the previous item is the last item in the last list
    # Note: the '> 1' is the part you'd modify to make it stricter or more relaxed
    if (x - lists[-1][-1]) / sd > 1:
        # then start a new list
        lists.append([])
    # add the current item to the last list in the list
    lists[-1].append(x)

print(lists)
```

```python
print(len(lists))
print("*"*100)
for li in lists:
    print(li)
```

```python
groups = []
for i in lists:
    if len(i)>1:
        x = str(i[0])
        diff_x = max_length - len(x)
        x=diff_x*str(0)+x
        
        y=str(i[-1])
        diff_y = max_length - len(y)
        y=diff_y*str(0)+y
        
        gr = x+"-"+y
        groups.append(gr)
        print(gr)
    else:
        x = str(i[0])
        diff_x = max_length - len(x)
        x=diff_x*str(0)+x
        gr = x
        groups.append(gr)
        print(gr)
```
